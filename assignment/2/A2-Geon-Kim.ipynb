**task 1: load dataset**

```python
import pandas as pd
import numpy as np

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
col = ['age', 'workclass','fnlwgt', 'education', 'education-num', 'martial-status','occupation', 
       'relationship','race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week','native-country', 'income']
incomes = pd.read_csv(url, sep=", ", names=col, engine='python')

print(incomes)
```

**task 2: looking into the dataset**

```python
print(incomes.head())
print(incomes.info())
print(incomes.describe())
print(incomes.shape)
```

**task 2.1: histogram**

```python
import matplotlib.pyplot as plt

incomes.hist(bins=20, figsize=(24, 16))
plt.show()
```

**task 3: missing values**

```python

incomes.isin(['?']).drop('income', axis=1).sum()
```

**task 3.1: replace '?' with NaN**

```python
incomes = incomes.replace('?', None)
incomes.isna().sum()
```

**task 4: duplicates**

```python
print(incomes.duplicated().sum())
incomes.drop_duplicates(inplace=True)
```
